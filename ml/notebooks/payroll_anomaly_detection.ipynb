# Payroll Anomaly Detection
This notebook demonstrates the feature engineering and model training workflow used by the automated pipeline.

```python
import pandas as pd
from pathlib import Path
from ml.scripts.feature_engineering import FeatureConfig, select_feature_columns
from ml.scripts.train_model import train
```

```python
config_path = Path("ml/config/training.yaml")
dataset = pd.read_csv("ml/data/curated_payroll_anomalies.csv")
dataset.head()
```

```python
feature_config = FeatureConfig(
    datetime_columns=["pay_period_start", "pay_period_end"],
    categorical_columns=["org_id", "employee_id", "employment_type"],
    numerical_columns=["gross_pay", "payg_withheld", "super_accrued"],
)
engineered = select_feature_columns(dataset, feature_config)
engineered.head()
```

```python
# Trigger a local training run and capture metrics
results = train(config_path)
results["metrics"]
```
